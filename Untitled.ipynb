```python
## Llei feble dels grans nombres aplicada al llancament d'una moneda n vegades
# on n es un enter variable que anirem augmentant una unitat a cada iteracio

import matplotlib.pyplot as plt
import random

n = 1       # variable que marcara el nombre de vegades que tirem moneda
prob = []   # cada element de l'array es la probabilitat de que surti cara per
            # un nombre de tirades n determinat
flip = []   # cada element de l'array es nombre de tirades n determinat

# Realitzarem 1000 estudis; Per cadascun d'ells, llancarem la moneda n vegades.
# Per exemple, si n = 20 ens trobarem en l'estudi numero 20 i llancarem moneda
# 20 vegades.
while n < 1000:
    # Per cada estudi reinicialitzem nombre de vegades que ha sortit cara o creu
    cara = 0
    creu = 0

    # Per cada tirada de moneda en l'estudi
    for i in range(n):
        # Si obtenim un 0, vol dir que ha sortit cara; ho augmentem a variable cara
        if random.randint(0, 1) == 0:
            cara += 1
        # Si no, vol dir que ha sortit creu (1); ho augmentem a variable creu
        else:
            creu += 1
    # Calculem la probabilitat de que surti cara en l'estudi n
    k = cara/(cara+creu)
    # Guardem la probabilitat de que surti cara en l'array prob
    prob.append(k)
    # Guardem el nombre de vegades que hem llancat moneda en l'array flip
    flip.append(n)
    # Fem estudi amb una tirada mÃ©s
    n += 1

# i els resultats els mapegem en un grafic

# subplot s'utilitza per indicar com sera el grafic de barres:
# subplot (2, 1, 1) --> 2 files, 1 columna, index on comenca el grafic; 1 per
# indicar cantonada inferior esquerra
plt.subplot(2, 1, 1)
# Generem un histograma de l'array prob (probabilitat de sortir cara) introduint
# l'array prob, el nombre de columnes que volem i el nom
plt.hist(prob, 100, label='Probabilitat de Cares')
# Mostrar la llegenda de l'histograma
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(flip, prob)
plt.xlabel("Nombre de Tirades")
plt.ylabel("Probabilitat de Cares")
plt.grid(True)

plt.show()
```
